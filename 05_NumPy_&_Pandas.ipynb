### **`NumPy Useful Fuctions:`**

```python
#2x3 array filled with 7
np.full((2,3), 7)
```

```python
#printing sum of elements of each rows separately
a6.sum(axis=1)
```

```python
#printing sum of elements of each columns separately
a7=a6.sum(axis=0)
a7
```

```python
#mean
np.mean(a7)
```

```python
#median
np.median(a6)
```

```python
#Standard Deviation
np.std(a6)
```

```python
#variance
np.var(a6)
```

```python
#minimum and maximum
print(f"Minimum: {np.min(a6)}")
print(f"Maximum: {np.max(a6)}")
```

```python
#Percentile

np.percentile(data, 50) #median
np.percentile(data, 25) #q1
np.percentile(data, 75) #q3
```
